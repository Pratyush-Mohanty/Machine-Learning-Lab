```python
from sklearn.metrics import accuracy_score
cov = [[0.8, 0, 0],
       [0, 0.8, 0],
       [0, 0, 0.8]]
class1_w1, class1_w2, class1_w3 = np.random.multivariate_normal([0, 0, 0], cov, 333).T
class2_w1, class2_w2, class2_w3 = np.random.multivariate_normal([1, 2, 2], cov, 333).T
class3_w1, class3_w2, class3_w3 = np.random.multivariate_normal([3, 3, 4], cov, 333).T

p_class1, p_class2, p_class3 = 1/3, 1/3, 1/3
```

```python
w1 = np.concatenate((class1_w1, class2_w1, class3_w1))
w2 = np.concatenate((class1_w2, class2_w2, class3_w2))
w3 = np.concatenate((class1_w3, class2_w3, class3_w3))

X = np.array([w1, w2, w3]).T
Y = [1]*333+[2]*333+[3]*333
```

```python
predictions = []
import math
for x in X:
    P_x_class1 = 1/(0.8*math.sqrt(2*math.pi))*np.exp(-0.5*((x-np.array([0, 0, 0]))/0.8)**2)
    P_x_class1 = P_x_class1[0]*P_x_class1[1]*P_x_class1[2]
    
    P_x_class2 = 1/(0.8*math.sqrt(2*math.pi))*np.exp(-0.5*((x-np.array([1, 2, 2]))/0.8)**2)
    P_x_class2 = P_x_class2[0]*P_x_class2[1]*P_x_class2[2]
    
    P_x_class3 = 1/(0.8*math.sqrt(2*math.pi))*np.exp(-0.5*((x-np.array([3, 3, 4]))/0.8)**2)
    P_x_class3 = P_x_class3[0]*P_x_class3[1]*P_x_class3[2]
    
    
    P_class1_x = P_x_class1*p_class1
    P_class2_x = P_x_class2*p_class2
    P_class3_x = P_x_class3*p_class3
    
    
    if P_class1_x >= P_class2_x >= P_class3_x or P_class1_x >= P_class3_x >= P_class2_x:
        predictions.append(1)
    elif P_class2_x >= P_class1_x >= P_class3_x or P_class2_x >= P_class3_x >= P_class1_x:
        predictions.append(2)
    else:
        predictions.append(3)
```

```python
print("Accuracy Score: ", round(accuracy_score(Y, predictions)*100, 2), "%")
```
